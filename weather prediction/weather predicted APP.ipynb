```python
import tkinter as tk
import pandas as pd
from sklearn.linear_model import LogisticRegression
from tkinter import messagebox
from PIL import Image, ImageTk


data = pd.read_csv('Appweather.csv')


X = data[['temp_max', 'temp_min', 'wind', 'precipitation']]
y = data['weather']


model = LogisticRegression()
model.fit(X, y)


window = tk.Tk()
window.title('Weather Prediction App')
window.geometry('400x400')
window.minsize(200,200)


bg_color = 'light blue'

def toggle_dark_mode():
    global bg_color
    if bg_color == 'light blue':
        bg_color = 'black'
        text_color = 'white'
    else:
        bg_color = 'light blue'
        text_color = 'black'
    window.configure(bg=bg_color)
    dark_mode_button.configure(bg=text_color, fg=bg_color)
    result_label.config(fg=text_color)
    temp_max_label.config(fg=text_color)
    temp_min_label.config(fg=text_color)
    wind_label.config(fg=text_color)
    precip_label.config(fg=text_color)
    temp_max_entry.config(fg=text_color, bg=bg_color)
    temp_min_entry.config(fg=text_color, bg=bg_color)
    wind_entry.config(fg=text_color, bg=bg_color)
    precip_entry.config(fg=text_color, bg=bg_color)

dark_mode_button = tk.Button(window, text='Dark Mode', command=toggle_dark_mode, bg='black', fg='white')
dark_mode_button.pack(anchor='ne')

def predict_weather():
    try:
        temp_max = float(temp_max_entry.get())
        temp_min = float(temp_min_entry.get())
        wind = float(wind_entry.get())
        precip = float(precip_entry.get())

        
        prediction = model.predict([[temp_max, temp_min, wind, precip]])

        
        result_label.config(text=f'Predicted Weather: {prediction[0]}')

       
        weather_icon_path = get_weather_icon(prediction[0])
        update_weather_icon(weather_icon_path)
    except ValueError:
        messagebox.showerror('Error', 'Please enter valid numeric values.')

def get_weather_icon(weather):
    
    icon_mapping = {
        'drizzle': 'drizzle.png',
        'rain': 'rain.png',
        'snow': 'snow.png',
        'sun': 'sun.png'
    }
    return icon_mapping.get(weather, 'unknown.png')

def update_weather_icon(icon_path):
    
    image = Image.open(icon_path)
    image = image.resize((100, 100), Image.ANTIALIAS)
    photo = ImageTk.PhotoImage(image)

    weather_icon_label.config(image=photo)
    weather_icon_label.image = photo


temp_max_label = tk.Label(window, text='Max Temperature (°F):', bg=bg_color, fg='black')
temp_max_label.pack()
temp_max_entry = tk.Entry(window)
temp_max_entry.pack()

temp_min_label = tk.Label(window, text='Min Temperature (°F):', bg=bg_color, fg='black')
temp_min_label.pack()
temp_min_entry = tk.Entry(window)
temp_min_entry.pack()

wind_label = tk.Label(window, text='Wind:', bg=bg_color, fg='black')
wind_label.pack()
wind_entry = tk.Entry(window)
wind_entry.pack()

precip_label = tk.Label(window, text='Precipitation (inches):', bg=bg_color, fg='black')
precip_label.pack()
precip_entry = tk.Entry(window)
precip_entry.pack()


predict_button = tk.Button(window, text='Predict Weather', command=predict_weather)
predict_button.pack()


result_label = tk.Label(window, text='', font=('Helvetica', 16), bg=bg_color, fg='black')
result_label.pack()

# Weather icon label
weather_icon_label = tk.Label(window, bg=bg_color)
weather_icon_label.pack()

# Start the GUI main loop
window.mainloop()
```
